```python
import pandas as pd
from sklearn.metrics import accuracy_score
import math
from dt import DecisionTree, dt_train_test, Node
```

```python
# load the train and test data
xTrain = pd.read_csv("q4xTrain.csv")
yTrain = pd.read_csv("q4yTrain.csv")
xTest = pd.read_csv("q4xTest.csv")
yTest = pd.read_csv("q4yTest.csv")
max_depth = 1
mls = [1, 2, 4, 8, 16, 32, 64, 128, 256]
columns= ['max_depth', "min_leaf_sample", "gini_train", 'gini_test', 'entropy_train', 'entropy_test']
```

```python
acc_output1 = pd.DataFrame(columns=columns)
acc_output1
```

```python
for min_leaf_sample in mls:
    dt1 = DecisionTree('gini', max_depth, min_leaf_sample)
    trainAcc1, testAcc1 = dt_train_test(dt1, xTrain, yTrain, xTest, yTest)
    print("gini   ", trainAcc1, testAcc1)
    dt = DecisionTree('entropy', max_depth, min_leaf_sample)
    trainAcc, testAcc = dt_train_test(dt, xTrain, yTrain, xTest, yTest)
    print("entropy", trainAcc, testAcc)

    new_row = {'max_depth':max_depth, "min_leaf_sample":min_leaf_sample, "gini_train":trainAcc1, 
               'gini_test':testAcc1, 'entropy_train':trainAcc, 'entropy_test':testAcc}
    # new_row = {'max_depth':2, "min_leaf_sample":1, "gini_train":2, 'gini_test':2,
    #            'entropy_train':2, 'entropy_test':2}
    acc_output1 = acc_output1.append(new_row, ignore_index=True)

acc_output1
```
