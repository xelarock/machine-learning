The first step is to import the necessary packages, datasets, and set the parameters that will be used

```python
import pandas as pd
from sklearn.metrics import accuracy_score
import math
from dt import DecisionTree, dt_train_test, Node
```

```python
# load the train and test data
xTrain = pd.read_csv("q4xTrain.csv")
yTrain = pd.read_csv("q4yTrain.csv")
xTest = pd.read_csv("q4xTest.csv")
yTest = pd.read_csv("q4yTest.csv")
max_depth = 1
mls = [1, 2, 4, 8, 16, 32, 64, 128, 256]
columns= ['max_depth', "min_leaf_sample", "gini_train", 'gini_test', 'entropy_train', 'entropy_test']
```

```python
acc_output1 = pd.DataFrame(columns=columns)                    # testing to see if I can create a blank dataframe
acc_output1                                                    # to store the values
```

I broke down running all the models because I was afraid that I'll lose all the data before it finished so by breaking it up, I could copy the values into dt-model-accuracy.csv to create the figures at a later time.

For the first set of accuracies, I test all the min leaf sample values on both entropy and gini criterion set at a fixed max depth of 1.

```python
for min_leaf_sample in mls:                                                 # for each value in min_leaf_sample
    dt1 = DecisionTree('gini', max_depth, min_leaf_sample)                  # make a DT for that set of parameters
    trainAcc1, testAcc1 = dt_train_test(dt1, xTrain, yTrain, xTest, yTest)  # get training and test accuracy.
    print("gini   ", trainAcc1, testAcc1)
    dt = DecisionTree('entropy', max_depth, min_leaf_sample)                # do the same for entropy 
    trainAcc, testAcc = dt_train_test(dt, xTrain, yTrain, xTest, yTest)
    print("entropy", trainAcc, testAcc)

    new_row = {'max_depth':max_depth, "min_leaf_sample":min_leaf_sample, "gini_train":trainAcc1, # create columns
               'gini_test':testAcc1, 'entropy_train':trainAcc, 'entropy_test':testAcc}
    acc_output1 = acc_output1.append(new_row, ignore_index=True)                            # add new row of data

acc_output1
```

Then, for this next chunk, I increased the max depth to 2 and again created and evaluated models on entropy and gini for all of the min leaf samples. 

```python
acc_output2 = pd.DataFrame(columns=columns)
acc_output2
max_depth = 2
for min_leaf_sample in mls:
    dt1 = DecisionTree('gini', max_depth, min_leaf_sample)
    trainAcc1, testAcc1 = dt_train_test(dt1, xTrain, yTrain, xTest, yTest)
    print("gini   ", trainAcc1, testAcc1)
    dt = DecisionTree('entropy', max_depth, min_leaf_sample)
    trainAcc, testAcc = dt_train_test(dt, xTrain, yTrain, xTest, yTest)
    print("entropy", trainAcc, testAcc)

    new_row = {'max_depth':max_depth, "min_leaf_sample":min_leaf_sample, "gini_train":trainAcc1, 
               'gini_test':testAcc1, 'entropy_train':trainAcc, 'entropy_test':testAcc}
    acc_output2 = acc_output2.append(new_row, ignore_index=True)

acc_output2
```
